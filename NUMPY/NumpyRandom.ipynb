```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#GENERATE RANDON FLOAT using random module rand()  returns random float 0 and 1
print("\n")
print("FLOAT")
x =random.rand()
print(x)

print("\n")
print("1-D ARRAY")
x = random.rand(5)
print(x)

print("\n")
print("2-D ARRAY")
x = random.rand(3,5)
print(x)
```

```python
#GENERATE A 1-D ARRAY CONTAINING 5 RANDOM INTEGERS FROM 0 TO 100
print("\n")
print("\n")
print("INTEGER")
print("1-D ARRAY")
x = random.randint(100, size=(5))
print(x)
```

```python
#GENERATE A 2-D ARRAY CONTAINING 5 RANDOM INTEGERS FROM 0 TO 100
print("\n")
print("2-D ARRAY")
x =random.randint(100, size=(3,5))
print(x)
```

```python
#GENERATE A RANDOM NUMBER FROM ARRAY
print("\n")
print("PRINTING A RANDOM NUMBER FROM A CREATED ARRAY")
x = random.choice([1,2,3,4,5,6])
print(x)
```

```python
#GENERATE A 2-D ARRAY THat CAN BE MADE BY THE CHOICE METHOD
print("\n")
print("2-D ARRAY BY choice method")
x = random.choice([1,2,3,4,5], size=(3,5))
print(x)
```

```python
#RANDOM DATA DISTRIBUTION
print("\n")
print("\n")
print("RANDOM DISTRIBUTION")
print("probability density function")
x = random.choice([1,2,3,4], p=[0.1, 0.3, 0.6, 0.0] , size=(100))
print(x)
```

```python
#RANDOM PERMUTATIONS 
print("\n")
print("Random permutation of elements")
print("\n")
print("SHUFFLING ARRAYS")
arr = np.array([1,2,3,4,5])
random.shuffle(arr)
print(arr)#SHUFFLING MAKES CHANGES TO THE ORIGINAL ARRAY
```

```python
#PERMUTATION OF ARRAYS
print("\n")
print("PERMUTATION OF ARRAYS")
arr = np.array([1,2,3,4,5])
random.permutation(arr)
print(arr)
```

```python
#SEABORN
#a library that uses Matplotlib underneath to plot graphs
print("\n")
print("\n")
print("VISUALIZATION WITH SEABORN")
sns.displot([0,1,2,3,4,5])#with histogram
plt.show()
```

```python
print("\n")
print("Without histogram")
sns.distplot([1,2,3,4,5], hist=False)
plt.show()
```

```python
#NORMAL(GAUSSIAN) DISTRIBUTION
print("\n")
print("\n")
print("NORMAL(GAUSSIAN) DISTRIBUTION")
x = random.normal(size=(2,3))#returns the shape of the array
print(x)

print("\n")
x = random.normal(loc=1, scale=2, size=(2,3))#returns mean, Standard Deviation and shape of the array
print(x)
```

```python
#BINOMIAL DISTRIBUTION 
print("\n")
print("\n")
print("BINOMIAL DISTRIBUTION")
x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

```python
#VISUALIZATION OF BINOMIAL DISTRIBUTION
print("\n")
print("visualization of binomial distribution")
sns.distplot(random.binomial(n=10, p=0.5, size=1000), kde=True, hist=False)
plt.show()
```

```python
#DIFFERENCE BETWEEN NORMAL AND BINOMIAL DISTRIBUTION
sns.distplot(random.normal(loc=50, scale=5, size=1000),
             hist=False, label='normal')#normal distribution
sns.distplot(random.binomial(n=100, p=0.5, size=1000),
             hist=False, label='binomial')

plt.show()
```

```python
#POISSON DISTRIBUTION 
print("\n")
print("\n")
print("POISSON DISTRIBUTION")
x = random.poisson(lam=2, size=10)
print(x)
```

```python
#VISUALIZATION OF POISSON DISTRIBUTION
print("\n")
print("\n")
print("visualization of poisson distribution")
sns.distplot(random.poisson(lam=4, size=1000), kde=False)
plt.show()
```

```python
#DIFFERENCE BETWEEN BINOMIAL AND POISSON DISTRIBUTION
sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.show()
```

```python
#DIFFERENCE BETWEEN BINOMIAL AND POISSON DISTRIBUTION
sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show()
```

```python
#Creating a 2x3 uniform distribution sample:
x = random.uniform(size=(2,3))
print(x)
```

```python
#visualization of uniform distribution
sns.distplot(random.uniform(size=1000),hist=False)
```

```python
#creating 2x3 samples from a logistic distribution with mean at 1 and stddev 2.0
x = random.logistic(loc=1, scale=2, size=(2,3))
print(x)
```

```python
#VISUALIZATION OF LOGISTIC DISTRIBUTION
sns.distplot(random.logistic(size=1000), hist=False)
plt.show()
```

```python
#LOGISTIC VS NORMAL DISTRIBUTION
sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')
plt.show()
```

```python
#MultiNomial Distribution
x = random.multinomial(n=6, pvals=[1/6,1/6,1/6,1/6,1/6,1/6])
print(x)#drawing out sample for dice roll
```

```python
#Exponential Distribution
x = random.exponential(scale=2, size=(2,3))
print(x)
```

```python
#visualization of exponential distribution
sns.distplot(random.exponential(size=1000),hist=False)
plt.show()
```

```python
#Chi Square distribution
#Draw out a sample for chi squared distribution with degree of freedom 2 with size 2x3:
x = random.chisquare(df=2, size=(2,3))
print(x)
```

```python
#visualization of the Chi Square distribution
sns.distplot(random.chisquare(df=1, size=1000), hist=False)
plt.show()
```

```python
#RAYLEIGH DISTRIBUTION 
x = random.rayleigh(scale=2, size=(2, 3))
print(x)
```

```python
#visualization of the rayleigh distribution
sns.distplot(random.rayleigh(size=1000), hist=False)
plt.show()
```

```python
#PARETO DISTRIBUTION
x = random.pareto(a=2, size=(2,3))
print(x)
```

```python
#VISUALIZATION OF PARETO DISTRIBUTION
sns.distplot(random.pareto(a=2, size=1000), kde=False)
plt.show()
```

```python
#ZIPF'S DISTRIBUTION

x = random.zipf(a=2, size=(2,3))
print(x)
```

```python
#Visualization of Zipf Distribution
x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)
plt.show()
```
